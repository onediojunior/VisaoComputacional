## **&#9776; VISÃO COMPUTACIONAL**
---



A visão computacional é um campo da inteligência artificial (IA) que permite que computadores e sistemas obtenham informações significativas de imagens digitais, vídeos e outras entradas visuais – e tomem ações ou façam recomendações com base nessas informações. Se a IA permite que os computadores pensem, a visão computacional permite que eles vejam, observem e compreendam.

A visão computacional precisa de muitos dados para realizar suas tarefas. Ele executa análises de dados repetidamente até discernir distinções e, finalmente, reconhecer imagens. Por exemplo, para treinar um computador para reconhecer pneus de automóveis, ele precisa ser alimentado com grandes quantidades de imagens de pneus e itens relacionados a pneus para aprender as diferenças e reconhecer um pneu, especialmente um sem defeitos.

Portanto, pelas definições acima, podemos dizer que a visão computacional é:
- a visão automatizada dos prcoessos visuais humanos;
- uma tarefa de processamento de informações;
- inversão da formação da imagem (Berthold Horn); e
- o inverso dos gráficos.

Duas tecnologias essenciais são usadas para fazer isso: um tipo de aprendizado de máquina chamado aprendizado profundo e uma rede neural convolucional (CNN).

## **&#x2609; Função resize()**

Função utilizada para redimensionar uma imagem para a forma desejada.

```python
# redimensionando a imagem para 313 por 313
# primeiro parâmetro - passar o objeto e o segundo o novo dimensionamento
imagem_alterada = cv2.resize(imagem,(315,313))
cv2_imshow(imagem_alterada)
```

##**&#9776; CONVERSÃO DE CORES**
---

Na conversão de cores, as imagens não são armazenadas com o método RGB (**R**ed, **B**lue, **G**reen) padrão, mas sim na ordem inversa, ou seja, na ordem **BGR**.

## **&#x2609; Função cvtColor()**

Função utilizada para modificar um código de cor para outro.

```python
# modifica código de cores
img_cinza = cv2.cvtColor(imagem_alterada, cv2.COLOR_BGR2GRAY)

# mostra a imagem modificada
cv2_imshow(img_cinza)
```

##**&#9776; DESENHOS**
---

Caso você necessite desenhar em sua imagem, seguem alguns argumentos que serão comuns:

- **img**: A imagem onde você deseja desenhar as formas
- **color**: Cor da forma. para BGR, passe como uma tupla, por exemplo: (255,0,0) para azul. Para tons de cinza, basta passar o valor escalar.
- **thickness**: Espessura da linha ou círculo etc. Se -1 for passado para figuras fechadas como círculos, ele preencherá a forma. Padrão thickness = 1.
- **lineType**: Tipo de linha, seja 8-connected, anti-aliased line, etc. Por padrão, é 8-connected. cv.LINE_AA fornece uma linha com suavização de serrilhado que parece ótima para curvas.

## **&#x2609; Função line()**

Para desenhar uma linha, você precisa passar as coordenadas inicial e final da linha.*texto em itálico*

```python
# desenhando na img_cinza duas linhas em branco (X)
nova_imagem = cv2.line(img_cinza,(60,60),(255,255),(255,255,255),5)
nova_imagem = cv2.line(nova_imagem,(255,60),(60,255),(255,255,255),5)
cv2_imshow(nova_imagem)
```

## **&#x2609; Função rectangle()**

Para desenhar um retângulo, você precisa do canto superior esquerdo e do canto inferior direito do retângulo.

```python
# desenhando na imagem_alterada um retângulo amarelo (X)
img_retangulo = cv2.rectangle(imagem_alterada,(20,20),(295,295),(0,255,255),5)
cv2_imshow(img_retangulo)
```

## **&#x2609; Função circle()**

Para desenhar um círculo, você precisa de suas coordenadas centrais e raio.

```python
# desenha um circulo amarelo vazado dentro da imagem_alterada
cv2_imshow(cv2.circle(img_retangulo,(157,155), 127, (0,255,255), 5))
```

##**&#9776;REGIÃO DE UMA IMAGEM (ROI)**
---

Para detecção de certas regiões de uma imagem, como os olhos, por exemplo, devemos detectar o rosto para posteriormente, buscarmos os olhos.

O processo é realizado dessa forma, pois melhora a precisão (porque os olhos estão sempre no rosto :D) e o desempenho (porque pesquisamos em uma área menor).

O ROI é obtido usando a indexação Numpy

```python
# aplicação de detecção em imagem
# com mudança de posição
img_futebol = io.imread("https://i.pinimg.com/originals/8d/43/ff/8d43ffcf11f2a9e1c5f47e5c2aaa6df0.jpg")
img_futebol = cv2.resize(img_futebol , (325,279))
cv2_imshow(img_futebol)

# captura bola
img_bola = img_futebol[45:107, 225:279]

# coloca a bola em outra posição
img_futebol[10:72, 40:94] = img_bola

# troca a cor dos pixels   
img_futebol[45:107, 225:279] = [246,246,246]

# apresenta a bola em nova posição
cv2_imshow(img_futebol)
```

## **&#x2609; Adição de imagem**

Você poderá adicionar duas imagens com a função **cv2.add()**, ou simplesmente pela operação Numpy adicao_img = img1 + img2.

Ambas as imagens devem ter a mesma profundidade e tipo, ou a segunda imagem pode ser apenas um valor escalar.

Cabe destacar que, a **adição do OpenCV** é uma **operação saturada**, enquanto a adição do **Numpy** é uma **operação de módulo**.

Portanto, é preferível sempre adotar as funções do **OpenCV**, porque elas fornecerão um resultado melhor.

```python
# aplica adição de duas imagens
bola_basquete = io.imread("https://www.freeiconspng.com/thumbs/basketball-png/basketball-png-7.png")
bola_basquete = cv2.resize(bola_basquete, (128,128))

bola_voley = io.imread("https://www.schoolofengineeringandsciences.org/sites/main/files/imagecache/lightbox/main-images/vbal1.png")
bola_voley = cv2.resize(bola_voley, (128,128))

img_adicao = cv2.add(bola_basquete, bola_voley)
cv2_imshow(img_adicao)
```

##**&#9776;MEDIÇÃO DE DESEMPENHO**

## **&#x2609; Função getTickCount()**

A função permite retornar o número de ciclos de clock após um evento de referência.

Então, se você chamá-lo antes e depois da execução da função, você obtém o número de ciclos de clock usados ​​para executar uma função.

## **&#x2609; Função getTickFrequency()**

A função retorna a frequência de ciclos de clock ou o número de ciclos de clock por segundo.

```python
# mede o desempenho para alterar uma imagem 
# para o tom de cinza e mostrar na tela

ev1 = cv2.getTickCount() 
cv2_imshow(cv2.cvtColor(img_adicao, cv2.COLOR_BGR2GRAY))
ev2 = cv2.getTickCount() 
tempo = (ev2 - ev1)/ cv2.getTickFrequency()
print(f" Tempo : {tempo}")
```
